```python
import time

# Python loop
start = time.time()
result = []
for i in range(1000000):
    result.append(i * 2)
print("Loop:", time.time() - start)

# NumPy vectorized
start = time.time()
arr = np.arange(1000000) * 2
print("NumPy:", time.time() - start)
```
